```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python

url_oscar = "https://es.wikipedia.org/wiki/Premios_%C3%93scar"
res_oscar = requests.get(url_oscar)
sopa_oscar = BeautifulSoup(res_oscar.content, "html.parser")
tabla_oscar = sopa_oscar.find("table", {"class": "wikitable"}) #en las tablas de wikipedia suelen tener "wikitable" como clase

if tabla_oscar:
    filas = tabla_oscar.find_all("tr") 
    lista_tuplas_oscars = []

    for fila in filas:
        celdas = fila.find_all("td")
        if celdas:
            tupla_oscars = (
                celdas[1].text.strip(),
                celdas[2].text.strip(),
                celdas[3].text.strip(),
                celdas[4].text.strip(),
                celdas[5].text.strip())
            # desde la celda 1 pq la 0 es una columna que no nos interesa. Los text.strip para sacar todos los espacios en cada celda que no nos interesan
            lista_tuplas_oscars.append(tupla_oscars)
            

    for tupla in lista_tuplas_oscars:
        print(tupla)
else:
    print("No se encontró la tabla de premios en la página.")
```
